```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
passengers_mean=np.mean(data['passengers'])
print(passengers_mean)
```

```python
sns.get_dataset_names()
data=sns.load_dataset('flights')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
plt.bar(data['year'],data['passengers'],label='PASSENGERS')
plt.axhline(y=passengers_mean,color='red',linestyle='--',label=f"Mean of passengers = {passengers_mean:.2f}")
plt.title('PASSENGERS VS YEARS')
plt.xlabel('YEARS')
plt.ylabel('PASSENGERS')
plt.legend()


plt.show()

```

```python
sns.get_dataset_names()
data=sns.load_dataset('flights')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
passengers_mean_every_year=data.groupby('year')['passengers'].mean()
plt.bar(passengers_mean_every_year.index,passengers_mean_every_year.values,label='PASSENGERS_MEAN_EVERY_YEAR')

plt.title('YEARS VS MEAN PASSENGERS')
plt.xlabel('YEARS')
plt.ylabel('MEAN PASSENGERS')
plt.legend()


plt.show()
```
