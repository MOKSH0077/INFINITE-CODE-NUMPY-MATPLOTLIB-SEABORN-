```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
data=sns.load_dataset('car_crashes')
data.head()
```

```python
crash_due_alcohol=np.mean(data.alcohol)
crash_due_speeding=np.mean(data.speeding)
crash_due_not_distracted=np.mean(data.not_distracted)
```

```python
print(crash_due_alcohol)
```

```python
plt.title("CAR_CRASH")
plt.plot(data.alcohol,label='ALCOHOL')
plt.axhline(y=crash_due_alcohol,color='blue',linestyle='--',label=f"Mean = {crash_due_alcohol:.2f}")
plt.plot(data.speeding,label='SPEEDING')
plt.axhline(y=crash_due_speeding, color='orange',linestyle='--',label=f"Mean = {crash_due_speeding:.2f}")
plt.plot(data.not_distracted,label='NOT_DISTRACTED')
plt.axhline(y=crash_due_not_distracted,color='green', linestyle='--', label=f"Mean = {crash_due_not_distracted:.2f}")
plt.xlabel("CAR_CRASH")
plt.ylabel("REASON")
plt.legend(loc='upper left')
plt.show()
```
