```python
product=pd.read_csv("products-100 (1).csv"
)
pd.set_option('display.max_columns',None)
product.head(5)
pr_mean=np.mean(product['Price'])
stck_mean=np.mean(product['Stock'])


plt.figure(figsize=(17,6))

plt.title("PRICE VS STOCK")
plt.plot(product['Brand'][:7],product['Price'][:7],label='PRICE',color='red')
plt.axhline(pr_mean,color='black',linestyle='--',label='MEAN PRICE')
plt.plot(product['Brand'][:7],product['Stock'][:7],label='STOCK',color='green')
plt.fill_between(product['Brand'][:7],product['Price'][:7],pr_mean,alpha=0.25,color='red',where=(product['Price'][:7]>pr_mean).to_numpy(),interpolate=True,label='PRICE>MEAN')

plt.legend()
plt.show()
```
