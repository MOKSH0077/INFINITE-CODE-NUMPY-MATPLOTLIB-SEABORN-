```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
expenses_in_oat=np.random.randint(1000,1200,size=10)


expenses_in_protein=np.random.randint(1200,1550,size=10)

expenses_in_vitamins=np.random.randint(1000,2000,size=10)

days=np.arange(1,11)
plt.title("EXPENSES ON SUPPLEMENTS FOR 10 DAYS IN RUPEES")
plt.plot(days,expenses_in_oat,label="OATS")
plt.plot(days,expenses_in_protein,label="PROTEIN")
plt.plot(days,expenses_in_vitamins,label="VITAMINS")
plt.xlabel("DAYS")
plt.ylabel("EXPENSES")

plt.legend()
plt.show()
```
