```python
prd=pd.read_csv("products.csv")
prd.head(10)
periods=[
datetime(2020,3,1),
datetime(2020,4,1),
datetime(2020,5,1),
datetime(2020,6,1),
datetime(2020,7,1),
datetime(2020,8,1),
datetime(2020,9,1),
datetime(2020,10,1),
datetime(2020,11,1),
datetime(2020,12,1),]

plt.figure(figsize=(12,6))
plt.xlabel("Month")
plt.ylabel("Price")
plt.title("PRICE OF SELP HELP BOOKS MONTH WISE")
plt.plot_date(periods,prd['Price'][:10],linestyle='--',label="ATOMIC HABIT")
plt.plot_date(periods,prd['Price'][10:20],linestyle='--',label='MEDITATION')
plt.plot_date(periods,prd['Price'][20:30],linestyle='--',label="RICH DAD POOR DAD")
plt.plot_date(periods,prd['Price'][30:40],linestyle='--',label="THE POWER OF NOW")
plt.gca().xaxis.set_major_formatter(mpl_dates.DateFormatter('%b, %Y'))
plt.gcf().autofmt_xdate()
plt.legend(loc="upper left")
plt.show()
```
